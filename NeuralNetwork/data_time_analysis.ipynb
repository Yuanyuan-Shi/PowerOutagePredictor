```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.tools.plotting import autocorrelation_plot
```

```python
# load data
data = pd.read_csv('../../Data/WeatherOutagesAllJerry.csv')
data = data.dropna(how = 'all')
X = range(0,data.shape[0])
Y = data['Total_outages']
```

```python
# plot how the number of outages change with time for one year
plt.figure(figsize=(10,5))
plt.plot(X[0:365],Y[0:365],color='red',label='total number of outages')
```

```python
# no significant time correlation for any time interval
autocorrelation_plot(Y[1:365])
plt.show()
```
